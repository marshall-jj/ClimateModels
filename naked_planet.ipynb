```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
time_step = 100 # years
water_depth = 4000 # meters
L = 1350 # solar radiation (W m^-2)
albedo = 0.3
```

```python
time_list = [0]
temperature_list = [0]

heat_capacity = water_depth * 4.2E6 # The amount of energy required to heat the surface 1 meter squared by 1 K (J)
heat_content = heat_capacity * temperature_list[0] # The sum of the system's internal energy.

heat_in = L * (1-albedo) / 4
heat_out = 0

def black_body(temp):
    epsilon = 1
    sigma = 5.67E-8 # W m^-2 K^4
    return epsilon * sigma * temp**4

for ts in range(0, 100):
    time_list.append(time_list[-1] + time_step)

    heat_out = black_body(temperature_list[-1])
    heat_content = heat_content + (heat_in - heat_out) * time_step * 3.14E7 # Increase in heat content over the period of the time step (3.14 is converting seconds in a year).
    
    temperature_list.append(heat_content / heat_capacity) # Appending the temperture for that time step.
```

```python
plt.figure(figsize=(20,10))
plt.scatter(time_list, temperature_list)

plt.title('Planet Surface Temperature as a Function of Time')
plt.ylabel('Temperature /K')
plt.xlabel('Time /years')

plt.show() 
```
